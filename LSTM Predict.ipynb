```python
!pip install keras-tuner
import os
import pandas as pd
import numpy as np
import time
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from kerastuner.tuners import RandomSearch
import matplotlib.pyplot as plt

# List of Excel files for multiple wells
well_files = ['57.xlsx', '197.xlsx', '218.xlsx', '265.xlsx', '269.xlsx', '341.xlsx']

# Combine data from multiple wells
combined_df = pd.DataFrame()

for file in well_files:
    df = pd.read_excel(file)
    df['VOLUME_DATE'] = pd.to_datetime(df['VOLUME_DATE'])
    df = df.set_index('VOLUME_DATE')

    # Select only the 'ALLOC_OIL_AWO' column
    df = df[['ALLOC_OIL_AWO']]

    # Aggregate data by mean
    aggregated_df = df.resample('D').mean()  # Resample data daily and calculate mean

    # Add a column for well label with "X-" prefix
    well_label = "X-" + file.split('.')[0]  # Add "X-" prefix to well label
    aggregated_df['WELL_LABEL'] = well_label

    # Combine the aggregated data
    combined_df = pd.concat([combined_df, aggregated_df], axis=0)

# Drop NaN values
combined_df = combined_df.dropna()

def create_sequences(data, seq_length):
    sequences = []
    targets = []
    for i in range(len(data) - seq_length):
        seq = data[i:i + seq_length]
        target = data[i + seq_length]
        sequences.append(seq)
        targets.append(target)
    return np.array(sequences), np.array(targets)

# Initialize lists to hold sequences and targets
X_train_list = []
y_train_list = []
X_test_list = []
y_test_list = []
scalers = {}

# Group by WELL_LABEL and create sequences for each group
sequence_length = 30
for label, group in combined_df.groupby('WELL_LABEL'):
    prod = group['ALLOC_OIL_AWO'].values.reshape(-1, 1)
    dates = group.index[sequence_length:]  # Skip the initial sequence length
    scaler = MinMaxScaler()
    production_data_normalized = scaler.fit_transform(prod)
    X, y = create_sequences(production_data_normalized, sequence_length)

    if len(X) > 0 and len(y) > 0:
        split = int(0.8 * len(X))
        X_train_temp, X_test_temp = X[:split], X[split:]
        y_train_temp, y_test_temp = y[:split], y[split:]
        dates_train, dates_test = dates[:split], dates[split:]

        X_train_list.append(X_train_temp)
        y_train_list.append(y_train_temp)
        X_test_list.append(X_test_temp)
        y_test_list.append(y_test_temp)
        scalers[label] = scaler

# Combine all wells data into single arrays
X_train = np.concatenate(X_train_list, axis=0)
y_train = np.concatenate(y_train_list, axis=0)
X_test = np.concatenate(X_test_list, axis=0)
y_test = np.concatenate(y_test_list, axis=0)

# Display shapes of the final arrays
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")

def build_model(hp):
    model = Sequential()
    model.add(LSTM(units=hp.Int('units', min_value=64, max_value=256, step=32), activation=hp.Choice('activation', values=['relu', 'tanh']), input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])), loss='mse', metrics=['mae'])
    return model

tuner = RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=5,
    executions_per_trial=3,
    directory='keras_tuner_logs',
    project_name='oil_production_forecasting'
)

tuner.search(X_train, y_train, epochs=10, validation_data=(X_test, y_test), batch_size=50, verbose=2)

# Get the best hyperparameters
best_hp = tuner.get_best_hyperparameters(num_trials=1)[0]
best_model = tuner.hypermodel.build(best_hp)

start = time.time()
# Train the best model
best_history = best_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100, batch_size=50, verbose=2)#, callbacks=[EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)])
end = time.time()
train_dur = end-start
print(f'Total Durasi Training Model: {train_dur} Seconds')

# Retrieve the best hyperparameters
print("Best Hyperparameters:")
best_hyperparameters = tuner.get_best_hyperparameters(num_trials=1)[0].values
for key, value in best_hyperparameters.items():
    print(f"{key}: {value}")

# Save the model
model_filename = 'best_model.h5'
best_model.save(model_filename)
print(f"Model saved as {model_filename}")

# Plot training history of the best model
plt.figure(figsize=(12, 6))
plt.plot(best_history.history['loss'], label='Training Loss')
plt.plot(best_history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss (Best Model)')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Initialize a DataFrame to store evaluation metrics for each well
evaluation_df = pd.DataFrame(columns=['Well', 'Root Mean Squared Error (Training)', 'Mean Absolute Percentage Error (Training)', 'R2 Score (Training)',
                                      'Root Mean Squared Error (WFV_OD)', 'Mean Absolute Percentage Error (WFV_OD)', 'R2 Score (WFV_OD)',
                                      'Root Mean Squared Error (WFV_Op)', 'Mean Absolute Percentage Error (WFV_Op)', 'R2 Score (WFV_Op)'])

# Plot predictions for each well with walk-forward validation and save to a folder
output_folder = "prediction_plots"
os.makedirs(output_folder, exist_ok=True)  # Create the output folder if it doesn't exist

for label, group in combined_df.groupby('WELL_LABEL'):
    scaler = scalers[label]

    prod = group['ALLOC_OIL_AWO'].values.reshape(-1, 1)
    dates = group.index[sequence_length:]  # Skip the initial sequence length
    production_data_normalized = scaler.transform(prod)
    X, y = create_sequences(production_data_normalized, sequence_length)

    split = int(0.8 * len(y))
    y_train_actual = scaler.inverse_transform(y[:split].reshape(-1, 1)).flatten()
    y_test_actual = scaler.inverse_transform(y[split:].reshape(-1, 1)).flatten()
    dates_train = dates[:split]
    dates_test = dates[split:]

    # Predict on the training set
    X_train_temp = X_train_list[list(combined_df.groupby('WELL_LABEL').groups.keys()).index(label)]
    y_pred_train_normalized = best_model.predict(X_train_temp)
    y_pred_train_actual = scaler.inverse_transform(y_pred_train_normalized).flatten()

    # Calculate APE for training data
    train_ape = abs((y_train_actual - y_pred_train_actual) / y_train_actual) * 100

    # Predict on the test set
    X_test_temp = X_test_list[list(combined_df.groupby('WELL_LABEL').groups.keys()).index(label)]
    y_pred_normalized = best_model.predict(X_test_temp)
    y_pred_actual = scaler.inverse_transform(y_pred_normalized).flatten()

    # Walk-forward validation for the test set using actual data
    wfv_preds_actual = []
    wfv_ape_actual = []
    walk_forward_dates = dates_test

    # Use only the test data for walk-forward validation
    X_test_for_wfv, y_test_for_wfv = X_test_temp, y[split:]
    for i in range(len(y_test_for_wfv)):
        seq = X_test_for_wfv[i].reshape(1, sequence_length, 1)
        pred = best_model.predict(seq)
        actual_value = scaler.inverse_transform(y_test_for_wfv[i].reshape(1, -1))[0][0]
        predicted_value = scaler.inverse_transform(pred)[0][0]
        wfv_preds_actual.append(predicted_value)
        wfv_ape_actual.append(abs((actual_value - predicted_value) / actual_value) * 100)

    # Walk-forward validation for the test set using predicted data
    wfv_preds_predicted = []
    wfv_ape_predicted = []
    current_sequence_predicted = X_test_temp[0]

    for i in range(len(y_test_actual)):
        seq = current_sequence_predicted.reshape(1, sequence_length, 1)
        pred = best_model.predict(seq)
        predicted_value = scaler.inverse_transform(pred)[0][0]
        wfv_preds_predicted.append(predicted_value)
        wfv_ape_predicted.append(abs((y_test_actual[i] - predicted_value) / y_test_actual[i]) * 100)

        # Update the sequence with the predicted value
        current_sequence_predicted = np.append(current_sequence_predicted[1:], pred)

    # Denormalize the data for plotting
    y_actual = scaler.inverse_transform(y.reshape(-1, 1)).flatten()

    # Plot train, test, and predicted data for each well
    plt.figure(figsize=(18, 7))
    #plt.plot(dates, y_actual, label=f"{label} - Actual", color='blue')
    plt.plot(dates[:split], y_actual[:split], label=f"{label} - Train", color='green')
    plt.plot(dates_test, y_test_actual, label=f"{label} - Test", color='red')
    plt.plot(dates_test, wfv_preds_actual, label=f"{label} - Walk-Forward Predicted (Actual Data)", color='orange', linestyle='dashed')
    plt.plot(dates_test, wfv_preds_predicted, label=f"{label} - Walk-Forward Predicted (Predicted Data)", color='purple', linestyle='dotted')
    plt.title(f'LSTM Predictions for {label}')
    plt.xlabel('Date')
    plt.ylabel('Production Volume')
    plt.legend()
    plt.savefig(os.path.join(output_folder, f'{label}_prediction_plot.png'))  # Save the plot as an image
    plt.close()

    # Plot APE for each well
    plt.figure(figsize=(18, 7))
    plt.plot(dates_train, train_ape, label=f"{label} - Training APE", color='blue', linestyle='solid')
    plt.plot(walk_forward_dates, wfv_ape_actual, label=f"{label} - Walk-Forward APE (Actual Data)", color='orange', linestyle='dashed')
    plt.plot(walk_forward_dates, wfv_ape_predicted, label=f"{label} - Walk-Forward APE (Predicted Data)", color='purple', linestyle='dotted')
    plt.axhline(y=10, color='r', linestyle='--', label='10% Error Line')
    plt.title(f'Absolute Percentage Error for {label}')
    plt.xlabel('Date')
    plt.ylabel('APE (%)')
    plt.legend()
    plt.savefig(os.path.join(output_folder, f'{label}_ape_plot.png'))  # Save the APE plot as an image
    plt.close()

    # Calculate evaluation metrics for the well
    rmse = np.sqrt(mean_squared_error(y_train_actual, y_pred_train_actual))
    mape = mean_absolute_percentage_error(y_train_actual, y_pred_train_actual)
    r2 = r2_score(y_train_actual, y_pred_train_actual)

    rmse_wfod = np.sqrt(mean_squared_error(y_test_actual, wfv_preds_actual))
    mape_wfod = mean_absolute_percentage_error(y_test_actual, wfv_preds_actual)
    r2_wfod = r2_score(y_test_actual, wfv_preds_actual)

    rmse_wfp = np.sqrt(mean_squared_error(y_test_actual, wfv_preds_predicted))
    mape_wfp = mean_absolute_percentage_error(y_test_actual, wfv_preds_predicted)
    r2_wfp = r2_score(y_test_actual, wfv_preds_predicted)
    # Append metrics to DataFrame
    evaluation_df = pd.concat([evaluation_df, pd.DataFrame([{
        'Well': label, 'Root Mean Squared Error (Training)': rmse, 'Mean Absolute Percentage Error (Training)': mape, 'R2 Score (Training)': r2,
                                      'Root Mean Squared Error (WFV_OD)': rmse_wfod, 'Mean Absolute Percentage Error (WFV_OD)': mape_wfod, 'R2 Score (WFV_OD)': r2_wfod,
                                      'Root Mean Squared Error (WFV_Op)': rmse_wfp, 'Mean Absolute Percentage Error (WFV_Op)': mape_wfp, 'R2 Score (WFV_Op)': r2_wfp

    }])], ignore_index=True)

# Save evaluation metrics to Excel
evaluation_df.to_excel(f"{output_folder}/well_evaluation_metrics.xlsx", index=False)
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Ensure plots directory exists
plots_dir = "plots"
if not os.path.exists(plots_dir):
    os.makedirs(plots_dir)

# Load the dataset
blind_test_file = '03.xlsx'
well_name = os.path.splitext(blind_test_file)[0]  # Extract the well name from the file name
blind_test_df = pd.read_excel(blind_test_file)
blind_test_df['Date'] = pd.to_datetime(blind_test_df['Date'])
blind_test_df['Date'] = blind_test_df['Date'].dt.date
blind_test_df = blind_test_df.groupby('Date', as_index=False).agg({'Oil': 'mean'})
blind_test_df = blind_test_df.set_index('Date')

# Normalize the data
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(blind_test_df['Oil'].values.reshape(-1, 1))

# Function to create sequences
def create_sequences(data, seq_length):
    sequences = []
    targets = []
    for i in range(len(data) - seq_length):
        seq = data[i:i + seq_length]
        target = data[i + seq_length]
        sequences.append(seq)
        targets.append(target)
    return np.array(sequences), np.array(targets)

# Sequence length
sequence_length = 166

# Create sequences for the entire dataset
X, y = create_sequences(data_normalized, sequence_length)

# Reshape for LSTM input
X = X.reshape((X.shape[0], X.shape[1], 1))

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=64, activation='tanh', input_shape=(X.shape[1], 1)))
model.add(Dropout(0.2))
model.add(Dense(1))
model.compile(optimizer=Adam(learning_rate=0.01), loss='mse', metrics=['mae'])

# Train the model
history = model.fit(X, y, epochs=100, batch_size=50, verbose=2)

# Walk-forward validation over actual data
actual_predictions = []

for i in range(sequence_length, len(data_normalized)):
    # Use actual data up to the current point to predict the next point
    seq = data_normalized[i-sequence_length:i].reshape(1, sequence_length, 1)
    pred = model.predict(seq)
    actual_predictions.append(pred[0][0])

# Invert normalization
actual_predictions_inv = scaler.inverse_transform(np.array(actual_predictions).reshape(-1, 1))
actual_targets_inv = scaler.inverse_transform(data_normalized[sequence_length:].reshape(-1, 1))

# Calculate metrics for actual data predictions
mae_actual = mean_absolute_error(actual_targets_inv, actual_predictions_inv)
rmse_actual = np.sqrt(mean_squared_error(actual_targets_inv, actual_predictions_inv))
mape_actual = mean_absolute_percentage_error(actual_targets_inv, actual_predictions_inv)
r2_actual = r2_score(actual_targets_inv, actual_predictions_inv)

print(f'WFV Actual Data - MAE: {mae_actual:.2f}, RMSE: {rmse_actual:.2f}, MAPE: {mape_actual * 100:.2f}%, R2: {r2_actual:.2f}')

# Walk-forward validation over predicted data
predicted_predictions = []

# Initialize with the first sequence of actual data
predicted_seq = data_normalized[:sequence_length].reshape(1, sequence_length, 1)

# Collect predictions iteratively
for i in range(sequence_length, len(data_normalized)):
    pred = model.predict(predicted_seq)
    predicted_predictions.append(pred[0][0])
    # Update the sequence by removing the first value and appending the predicted value
    predicted_seq = np.append(predicted_seq[:, 1:, :], [[pred[0]]], axis=1)

# Invert normalization
predicted_predictions_inv = scaler.inverse_transform(np.array(predicted_predictions).reshape(-1, 1))
predicted_targets_inv = scaler.inverse_transform(data_normalized[sequence_length:].reshape(-1, 1))

# Calculate metrics for predicted data predictions
mae_predicted = mean_absolute_error(predicted_targets_inv, predicted_predictions_inv)
rmse_predicted = np.sqrt(mean_squared_error(predicted_targets_inv, predicted_predictions_inv))
mape_predicted = mean_absolute_percentage_error(predicted_targets_inv, predicted_predictions_inv)
r2_predicted = r2_score(predicted_targets_inv, predicted_predictions_inv)

print(f'WFV Predicted Data - MAE: {mae_predicted:.2f}, RMSE: {rmse_predicted:.2f}, MAPE: {mape_predicted * 100:.2f}%, R2: {r2_predicted:.2f}')

# Calculate absolute percentage error for walk-forward validation
ape_actual_wfv = np.abs((actual_targets_inv - actual_predictions_inv) / actual_targets_inv) * 100
ape_predicted_wfv = np.abs((predicted_targets_inv - predicted_predictions_inv) / predicted_targets_inv) * 100

# Plot Actual vs. Predicted for Walk-forward Validation
plt.figure(figsize=(18, 6))
plt.plot(blind_test_df.index, blind_test_df['Oil'], label='Actual')
plt.plot(blind_test_df.index[-len(actual_predictions_inv):], actual_predictions_inv, label='WFV (Actual Data)')
plt.plot(blind_test_df.index[-len(predicted_predictions_inv):], predicted_predictions_inv, label='WFV (Predicted Data)')
plt.title(f'LSTM Blind Test Well X-{well_name}')
plt.xlabel('Date')
plt.ylabel('Production Volume')
plt.legend()
plt.savefig(os.path.join(plots_dir, f'walk_forward_validation_{well_name}.png'))
plt.show()

# Plot absolute percentage error for walk-forward validation over actual and predicted data
plt.figure(figsize=(18, 6))
plt.plot(blind_test_df.index[-len(ape_actual_wfv):], ape_actual_wfv, label='APE (WFV Actual Data)')
plt.plot(blind_test_df.index[-len(ape_predicted_wfv):], ape_predicted_wfv, label='APE (WFV Predicted Data)')
plt.axhline(y=10, color='r', linestyle='--', label='10% Error Line')
plt.title(f'APE for Walk-forward Validation Over Time for Well {well_name}')
plt.xlabel('Date')
plt.ylabel('Absolute Percentage Error (%)')
plt.legend()
plt.savefig(os.path.join(plots_dir, f'ape_wfv_{well_name}.png'))
plt.show()

# Save evaluation metrics to Excel
evaluation_metrics = pd.DataFrame({
    'Metric': ['WFV Actual MAE', 'WFV Actual RMSE', 'WFV Actual MAPE', 'WFV Actual R2', 'WFV Predicted MAE', 'WFV Predicted RMSE', 'WFV Predicted MAPE', 'WFV Predicted R2'],
    'Value': [mae_actual, rmse_actual, mape_actual, r2_actual, mae_predicted, rmse_predicted, mape_predicted, r2_predicted]
})

evaluation_metrics.to_excel(f"{plots_dir}/evaluation_metrics_{well_name}.xlsx", index=False)
print(evaluation_metrics)
```
